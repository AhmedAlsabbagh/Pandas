```python
series_eg4 = pd.Series([10,30,30,40,50],index=['d', 'a', 'b', 'c','e'])
print(series_eg4)

print(series_eg4.rank())                                                        # Default is average

print(series_eg4.rank(method='first'))                                          # Lowest value will get rank 1

print(series_eg4.rank(ascending=False, method='first'))                         # Highest value will get rank 1

print(series_eg4.rank( method='max'))                                           # Tie breaking, assign max rank to tie
```

```python
# Create series from Dictionary
exp = [1, 0, 2, 3, 4, 5, 6, 7]                                                  # Define a list
salary = {0:8, 1:9, 2:7, 3:8, 4:9, 5:11, 6:15, 6:14, 8:58}                      # Define a dictionary. Note that 6 key is repeated and will take latest value 14

exp_salary = pd.Series(salary,index=exp)                                        # Create a series with labels as "exp" and value from "salary"
exp_salary

#Q1. why 7 is NaN? A1: Because 7 was present in index=exp list but not present in values=salary dictionary
#Q2. Why 6 got value as 14, but not 15,  (6:15, 6:14) A2. Since salary is a dictionary and it can not have duplicate keys, it just retains the latest value , 14 in this case
#Q3. Why 8:58 is missing? A3. Because 8 is not present in the index=exp list
```

```python
# Apply a function to dataframe elements => Column wise and Row Wise
df5 = pd.DataFrame(np.arange(9.).reshape((3, 3)),  index=['r1', 'r2', 'r3'], columns=list('abc'))
print(df5)

func = lambda x: x.max() - x.min()

print(df5.apply(func,axis='index'))                                                          # Diff between max and min element for each column
print(df5.apply(func, axis = 'columns'))                                        # Diff between max and min element for each row
```

```python
# Read a csv and store into file separated by "$"
# Also if marks is 32.0, store marks as 35.0
import csv

# Step 1: Define Dialect for Read and open file to be read
class customReadDialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ','
    quotechar = '"'
    quoting = csv.QUOTE_MINIMAL                                                 # Quote only fields with special characters like delimiters

fileToRead = open('temp210823_noheader.csv')
customReader = csv.reader(fileToRead, dialect=customReadDialect)

# Step 2: Define Dialect for Write and open file to be Writen
class customWriteDialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = '$'
    quotechar = '"'
    quoting = csv.QUOTE_MINIMAL

fileToWrite = open('temp210823_toWrite.csv', 'w')
customWriter = csv.writer(fileToWrite, dialect=customWriteDialect)

# Step 3: Read line by line and write
for line in customReader:
  if  line[5] == '32.0':
      line[5] = '35.0'
  customWriter.writerow(line)
```

1. Stack

```python
student_marks_df = pd.DataFrame({
    "Subject": ['English','Maths','Science'],
    "marks": [22,21,24]
}, index = ["personA","personB", "personC"])
print(student_marks_df)

print('\n On Stack')
print(student_marks_df.stack())                                                 # Returns series from dataframe, by bringing column name to index
print(type(student_marks_df.stack()))

print('\n On Unstack')
print(student_marks_df.stack().unstack())
print(type(student_marks_df.stack().unstack()))                                 # Returns back dataframe
```

```python
print('\n On Unstack')
print(student_marks_df.stack().unstack(0))                                      # For unstack you can specify which level to unstack
```

```python
student_marks_repeated_df = pd.DataFrame({                                      # With duplicate data
    "name": ["personA","personA","personA","personB","personB","personB","personC","personC","personC","personA","personA","personA","personB","personB","personB",],
    "subject": ["English", "Maths", "Science","English", "Maths", "Science","English", "Maths", "Science","English", "Maths", "Science","English", "Maths", "Science"],
    "marks": [22,21,24,20,23,24,21,22,22,23,22,25,21,24,25],
    "grade": ["A","B","A","B","A","A","B","A","A","A","B","A","B","A","A"]
})
student_marks_repeated_df
```

```python
#student_marks_repeated_df.pivot(index='subject',columns='name', values='marks') # Cannot handle duplicates, see pivot_table
```

```python
aggregated_df = student_marks_repeated_df.pivot_table(index='subject',columns='name',values='marks', aggfunc=np.sum)
                                                                                # handles duplicate, default is aggfunc=np.mean
aggregated_df
```

```python
aggregated_df = student_marks_repeated_df.pivot_table(index='subject',columns='name',values='marks', aggfunc=[np.sum, 'count'])
                                                                                # Can write multiple functions to perform,  aggfunc=np.mean, 'count'
aggregated_df
```

```python
aggregated_df = student_marks_repeated_df.pivot_table(index='subject',columns='name',values='marks', aggfunc=np.sum, margins=True)
                                                                                # "margins=True" margin will add partial totals
aggregated_df
```

```python
# crosstab - Same as pivot_table with aggregation fucntion count
aggregated_df = pd.crosstab(student_marks_repeated_df.subject,student_marks_repeated_df.name, margins=True)
                                                                                # crosstab is simmilar to "pivot_table with count function" - see above pivot_table with count
aggregated_df
```
